## Fungsi rekomendasi berbasis film

```python
indices = pd.Series(movie_clean.index, index=movie_clean['title'])
print(indices['Toy Story (1995)'])
```

```python
idx = indices['Toy Story (1995)']
print(cosine_sim[idx])
```

```python
sim_scores = list(enumerate(cosine_sim[idx]))
sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse = True) 
print(sim_scores)
```

```python
def get_recomendations(title, n=10):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse = True)
    sim_scores = sim_scores[1:n+1]
    movie_indices = [i[0] for i in sim_scores]
    return movie_clean.iloc[movie_indices][['title', 'genres']]    
```

# Implementasi SVD

```python
from surprise import SVD, Dataset, Reader
from surprise.model_selection import train_test_split, cross_validate
```

```python
reader = Reader(rating_scale=(0.5, 5.0))
```

```python
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)
```

```python
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)
```

```python
model = SVD(n_factors=100, random_state=42)
model.fit(trainset)
```

```python
predictions = model.test(testset)
```

```python
from surprise import accuracy
```

```python
rmse = accuracy.rmse(predictions)
mae = accuracy.mae(predictions)
```

```python
def svd_recomendations(user_id, n=10):
    watched_movies = ratings[ratings['userId'] == user_id]['movieId'].tolist()
    unwatched_movies = movie_clean[~movie_clean['movieId'].isin(watched_movies)]
    predictions = []
    for movie_id in unwatched_movies['movieId']:
        pred = model.predict(uid=user_id, iid=movie_id)
        predictions.append((movie_id, pred.est))
    predictions = sorted(predictions, key=lambda x: x[1], reverse=True)
    top_movies = [i[0] for i in predictions[:n]]
    return movie_clean[movie_clean['movieId'].isin(top_movies)][['title', 'genres']]
```
